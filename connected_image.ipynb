```python
import numpy as np
from skimage import measure, color
import matplotlib.pyplot as plt

def component_labeling(image_array):
    """
    Performs component labeling on a binary image and visualizes the result.

    Args:
        image_array (numpy.ndarray): A 2D NumPy array representing the image,
                                     where 0 is black (object) and 255 is white (background).
    """

    # Ensure the image is binary (0 or 1) for labeling
    binary_image = (image_array == 0).astype(int)

    # Perform connected component labeling
    labeled_image = measure.label(binary_image, connectivity=2)  # 2 for 4-connectivity

    # Visualize the labeled image using label2rgb
    image_label_overlay = color.label2rgb(labeled_image, image=image_array, bg_label=0)

    # Display the original image and the labeled image
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(image_array, cmap='gray')
    axes[0].set_title('Original Image')
    axes[1].imshow(image_label_overlay)
    axes[1].set_title('Labeled Components')
    plt.show()

# Example usage (replace with your image data)
image_data = np.array([
    [255, 255, 0, 0, 255, 255, 255, 255, 255, 255],
    [255, 0, 0, 255, 255, 255, 255, 255, 255, 255],
    [255, 255, 255, 255, 255, 255, 0, 0, 0, 255],
    [255, 255, 255, 255, 255, 255, 0, 0, 255, 255],
    [255, 255, 255, 255, 0, 0, 255, 255, 255, 255],
    [0, 0, 255, 255, 255, 255, 255, 255, 255, 255],
    [0, 0, 0, 255, 255, 255, 255, 255, 255, 255],
    [0, 0, 0, 0, 0, 255, 0, 0, 255, 255],
    [255, 255, 255, 0, 0, 0, 255, 255, 255, 255],
    [255, 255, 255, 255, 255, 255, 255, 255, 255, 255]
])

component_labeling(image_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import measure, color

# Sample binary image (0 for region and 255 for background)
image = np.array([
    [255, 255, 0, 0, 255, 255, 255, 255, 255, 255],
    [255, 0, 0, 255, 255, 255, 255, 255, 255, 255],
    [255, 255, 255, 255, 255, 255, 0, 0, 0, 255],
    [255, 255, 255, 255, 255, 255, 0, 0, 255, 255],
    [255, 255, 255, 255, 0, 0, 255, 255, 255, 255],
    [0, 0, 255, 255, 255, 255, 255, 255, 255, 255],
    [0, 0, 0, 255, 255, 255, 255, 255, 255, 255],
    [0, 0, 0, 0, 0, 255, 0, 0, 255, 255],
    [255, 255, 255, 0, 0, 0, 255, 255, 255, 255],
    [255, 255, 255, 255, 255, 255, 255, 255, 255, 255]
])


# connected components labeling is done on regions with value 0 (black region).
# connectivity=2 for 8-connected components (horizontal, vertical, or diagonal neighbors).
label_image = measure.label(image == 0, connectivity=2)

# add colors with labels
labeled_image = color.label2rgb(label_image, bg_label=255)

#Original image
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original Binary Image")
ax[0].axis('off')

# Labeled image 
ax[1].imshow(labeled_image)
ax[1].set_title("Labeled Components (Pseudo-color)")
ax[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import measure, color

# Sample binary image (0 for region and 255 for background)
image = np.array([
    [255, 255, 0, 0, 255, 255, 255, 255, 255, 255],
    [255, 0, 0, 255, 255, 255, 255, 255, 255, 255],
    [255, 255, 0, 0, 255, 255, 255, 255, 255, 255],
    [255, 255, 255, 255, 255, 255, 255, 255, 255, 255],
    [255, 255, 255, 255, 255, 255, 255, 255, 255, 255]
], dtype=np.uint8)

# Step 1: Perform component labeling
# In skimage, connected components labeling is done on regions with value 0 (black region).
# Use connectivity=4 for 4-connected components (horizontal/vertical neighbors).
label_image = measure.label(image == 0, connectivity=2)

# Step 2: Convert the labeled image into a pseudo-colored image
# Here, bg_label=255 ensures the background is shown as white.
labeled_image = color.label2rgb(label_image, bg_label=255)

# Step 3: Display the original image and the labeled pseudo-color image
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Original binary image (showing the black region and white background)
ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original Binary Image")
ax[0].axis('off')

# Labeled image with pseudo-color (regions labeled with different colors)
ax[1].imshow(labeled_image)
ax[1].set_title("Labeled Components (Pseudo-color)")
ax[1].axis('off')

plt.tight_layout()
plt.show()
```
